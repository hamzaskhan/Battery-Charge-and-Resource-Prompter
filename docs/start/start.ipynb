```
import cirq
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_qubits = 5
qubits = [cirq.LineQubit(i) for i in range(num_qubits)]
lambda_c = 3  # Threshold for dimensional regression
exponential_growth_factor = 3  # Controls exponential growth of entangling gates
num_cycles = 9  # Number of cycles to run

# Initialize storage for entropy and measurement probabilities
entropy_values = []
probability_values = []

# Function to calculate von Neumann entropy
def calculate_entropy(state_vector, subsystem_qubits):
    """
    Calculate the von Neumann entropy for the specified subsystem of qubits.
    """
    # Create density matrix from state vector for specified qubits
    density_matrix = cirq.qis.density_matrix_from_state_vector(state_vector, [q.x for q in subsystem_qubits])
    eigenvalues = np.linalg.eigvalsh(density_matrix)
    eigenvalues = eigenvalues[eigenvalues > 0]  # Filter out zero eigenvalues to avoid log(0)
    entropy = -np.sum(eigenvalues * np.log2(eigenvalues))
    return entropy

# Function to create a GHZ-like state with exponential entanglement growth
def create_ghz_circuit(num_entangling_gates):
    """
    Create the circuit with GHZ-like state preparation and entangling gates.
    """
    circuit = cirq.Circuit()

    # Prepare GHZ state (maximally entangled)
    circuit.append(cirq.H(qubits[0]))  # Hadamard gate on the first qubit
    for i in range(1, num_qubits):
        circuit.append(cirq.CNOT(qubits[0], qubits[i]))  # CNOT entangling with qubit 0

    # Add additional entangling gates if needed (based on the exponential growth factor)
    for i in range(1, num_qubits):
        for _ in range(num_entangling_gates):
            circuit.append(cirq.CNOT(qubits[i - 1], qubits[i]))  # Entangling gates

    # Add measurements to the circuit
    circuit.append(cirq.measure(*qubits, key='result'))  # Measure all qubits

    return circuit

# Initialize the Cirq simulator
simulator = cirq.Simulator()

# Run experiment and calculate entropy with dynamic entangling gate count
for cycle in range(num_cycles):
    # Calculate number of entangling gates with exponential growth
    num_entangling_gates = int(np.round(exponential_growth_factor ** cycle))

    # Create and run the GHZ-like entangling circuit
    circuit = create_ghz_circuit(num_entangling_gates)

    # Simulate the final state vector
    state_vector = simulator.simulate(circuit).final_state_vector
    entropy = calculate_entropy(state_vector, qubits[:num_qubits//2])  # Measure entropy of the first half
    entropy_values.append(entropy)

    # Track probability distribution of measurement outcomes
    result = simulator.run(circuit, repetitions=1000)
    measurement_counts = result.histogram(key='result')

    # Track probabilities for each possible outcome
    total_measurements = sum(measurement_counts.values())
    probabilities = {outcome: count / total_measurements for outcome, count in measurement_counts.items()}
    probability_values.append(probabilities)

    # For visualization: Display the current state vector as a heatmap
    plt.figure(figsize=(6, 6))
    plt.imshow(np.abs(state_vector).reshape((1, -1)), cmap='plasma', aspect='auto')
    plt.colorbar(label="Amplitude Magnitude")
    plt.title(f"State Vector Amplitude Heatmap - Cycle {cycle}")
    plt.xlabel("Qubit Index")
    plt.ylabel("State Vector Amplitude")
    plt.show()

# Plot Entropy Over Time
plt.figure(figsize=(10, 5))
plt.plot(range(num_cycles), entropy_values, color='green', linestyle='-', linewidth=2)
plt.xlabel('Cycle')
plt.ylabel('Entanglement Entropy')
plt.title('Entropy Growth Over Time: Simulating Big Bang Expansion')
plt.show()

# Plot Probability Distribution Over Time
plt.figure(figsize=(10, 5))
for cycle, probs in enumerate(probability_values):
    x = list(probs.keys())
    y = list(probs.values())
    plt.plot(x, y, label=f'Cycle {cycle}')

plt.xlabel('Measurement Outcome')
plt.ylabel('Probability')
plt.title('Measurement Outcome Probabilities Over Time')
plt.legend()
plt.show()
```

```
import cirq
import numpy as np

# Define the entropy computation function before using it
def compute_entropy(result):
    """
    Computes Von Neumann entropy from the measurement results.
    Uses the histogram of measurements to calculate entropy.
    """
    # Get the measurement counts
    counts = result.histogram(key='result')  # Use the 'result' key to access measurement outcomes

    # Normalize the counts to get probabilities
    probabilities = np.array([count / sum(counts.values()) for count in counts.values()])

    # Calculate entropy using the formula: H = -Σ(p * log2(p))
    entropy = -np.sum(probabilities * np.log2(probabilities))
    return entropy

# Function for the expansion phase of the system
def expansion_phase(circuit, qubits, lambda_val):
    """
    This function simulates the expansion of dimensions in the system by applying
    CNOT and Rx gates, followed by a measurement to capture the quantum state.
    """
    # Apply CNOT gates as part of dimensional expansion (creating entanglement)
    for i in range(len(qubits)-1):
        circuit.append(cirq.CNOT(qubits[i], qubits[i+1]))  # Creating entanglement

    # Gradually increase Lambda by applying Rx gates to simulate dimensional growth
    for i in range(len(qubits)):
        circuit.append(cirq.rx(lambda_val)(qubits[i]))  # Rotate qubits to simulate dimensional growth

    # Add a measurement to all qubits to capture the quantum state and enable entropy calculation
    circuit.append(cirq.measure(*qubits, key='result'))  # Measure all qubits

# Simulate the expansion phase for increasing Lambda
lambda_vals = np.linspace(0.1, 1.0, 50)  # Lambda range from 0.1 to 1.0
expansion_entropies = []

# List of qubits to simulate on
qubits = [cirq.NamedQubit(f'q{i}') for i in range(3)]  # Example: 3 qubits

for lambda_val in lambda_vals:
    # Copy the initial circuit and apply the expansion phase
    circuit = cirq.Circuit(cirq.H(q) for q in qubits)  # Starting circuit with Hadamard gates for superposition
    circuit_expansion = cirq.Circuit(circuit)  # Copy the initial circuit
    expansion_phase(circuit_expansion, qubits, lambda_val)

    # Simulate the quantum circuit and collect the measurement results
    result = cirq.Simulator().run(circuit_expansion, repetitions=100)

    # Compute entropy from the measurement results
    entropy = compute_entropy(result)

    # Store the entropy for this lambda value
    expansion_entropies.append(entropy)

# Display the results (optional)
import matplotlib.pyplot as plt

plt.plot(lambda_vals, expansion_entropies)
plt.xlabel('Lambda Value')
plt.ylabel('Entropy')
plt.title('Entropy vs Lambda during Dimensional Expansion')
plt.show()
```

```
def collapse_phase(circuit, qubits, lambda_val):
    # Apply reverse CNOT gates (reverse of expansion)
    for i in range(len(qubits)-1, 0, -1):
        circuit.append(cirq.CNOT(qubits[i-1], qubits[i]))  # Reverse the CNOT gates

    # Apply reverse rotations to simulate dimensional collapse
    for i in range(len(qubits)):
        circuit.append(cirq.rx(-lambda_val)(qubits[i]))  # Reverse the rotations

    # Add a measurement to all qubits to capture the quantum state and enable entropy calculation
    circuit.append(cirq.measure(*qubits, key='result'))  # Measure all qubits

# Simulate the collapse phase for decreasing Lambda
collapse_entropies = []

for lambda_val in reversed(lambda_vals):
    circuit_collapse = cirq.Circuit(circuit)  # Copy the initial circuit
    collapse_phase(circuit_collapse, qubits, lambda_val)

    # Simulate the quantum circuit and collect the measurement results
    result = cirq.Simulator().run(circuit_collapse, repetitions=100)

    # Compute entropy from the measurement results
    entropy = compute_entropy(result)

    # Store the entropy for this lambda value
    collapse_entropies.append(entropy)
```
